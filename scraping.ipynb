```python
def scrape(nom,prenom):
    from PIL import Image
    import re
    import requests
    from bs4 import BeautifulSoup
    if (isinstance(nom,str) == True) and (isinstance(prenom,str) == True):
        url = 'https://en.wikipedia.org/wiki/'+nom+"_"+prenom
        soup = BeautifulSoup(requests.get(url).text,'html.parser')
        imglinks = soup.find_all('a', attrs = {'class':'image'})[0]
        for img in imglinks.find_all('img'):
            img['src'].replace('//','https://')
            headers = {
            'User-Agent': 'My User Agent 1.0'
            }
            picture_url = img['src'].replace('//','https://')
            r = requests.get(picture_url, headers=headers, stream=True)
            return Image.open(r.raw)
    else:
        print("Veuillez entrez un nom et pr√©nom valables")
```
